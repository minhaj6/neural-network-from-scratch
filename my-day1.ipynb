```python
import sys
import numpy as np
import matplotlib
```

```python
# making a neuron with three inputs

inputs = [1, 2, 3]      # 3 inputs coming from 3 previous neurons
weights = [0.2, 0.8, -0.5]    # every input has a weight associated
bias  = 2                 # every neuron has a unique bias

output = bias
for i,input in enumerate(inputs):
  output += (input * weights[i])

print(output)      # output = sum(input * weight) + bias
```

```python
# model 3 neurons with 4 inputs
# - 3 separate weight sets, each set has 4 values
# - 3 biases

inputs = [1, 2, 3, 2.5]

weights1 = [0.2, 0.8, -0.5, 1.0]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]

bias1 = 2
bias2 = 3
bias3 = 0.5

output = [bias1, bias2, bias3]
for i,input in enumerate(inputs):
  output[0] += (input * weights1[i])
  output[1] += (input * weights2[i])
  output[2] += (input * weights3[i])

print(output)
  
```

```python
# make the previous code more dynamic

inputs =  [1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

layer_outputs = []
for neuron_weights, neuron_bias in zip(weights, biases):
  neuron_output = 0
  for input, weight in zip(inputs, neuron_weights):
    neuron_output += input * weight
  neuron_output += neuron_bias
  layer_outputs.append(neuron_output)

print(layer_outputs)

# can be done with numpy
# output = np.dot(weight, input) + bias    ... weight should be first parameter *** 
```

```python
inputs =  [1, 2, 3, 2.5]
weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]

outputs = np.dot(weights, inputs) + biases
print(outputs)
```
